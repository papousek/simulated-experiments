```python
import matplotlib.pyplot as plt
import numpy
import numpy.random
import random
from pprint import pprint
%autoreload 2
from proso.model import *
from proso.simulator import Simulator
from proso.util import *
```

```python
import seaborn as sns
COLORS = sns.color_palette()
from IPython.core.pylabtools import figsize
figsize(11, 9)
```

```python
NUMBER_OF_ITEMS = 200
NUMBER_OF_ITEMS_WITH_BAD_CLUSTER = 50
NUMBER_OF_USERS = 1000
NUMBER_OF_CLUSTERS = 2
PRACTICE_LENGTH = 50
TARGET_PROBABILITY = 0.75
```

```python
items = dict(map(lambda i: (i, numpy.random.normal(0, 1)), range(NUMBER_OF_ITEMS)))
clusters = dict(map(lambda i: (i, random.choice(range(NUMBER_OF_CLUSTERS))), range(NUMBER_OF_ITEMS)))
users = dict(map(lambda u: (u, numpy.random.normal(0, 1, size=NUMBER_OF_CLUSTERS)), range(NUMBER_OF_USERS)))
```

```python
optimal_model = OptimalModel(users, items, clusters)
optimal_simulator = Simulator(optimal_model, optimal_model, users, items, clusters, practice_length=PRACTICE_LENGTH)
optimal_simulator.simulate()
optimal_simulator.simulate_all()
```

```python
elo_model = ClusterEloModel(clusters={})
elo_simulator = Simulator(optimal_model, elo_model, users, items, clusters, practice_length=PRACTICE_LENGTH)
elo_simulator.simulate()
elo_simulator.simulate_all()
```

```python
wrong_items = random.sample(items.keys(), NUMBER_OF_ITEMS_WITH_BAD_CLUSTER)
wrong_clusters = dict(clusters.items())
for i in wrong_items:
    available_clusters = set(range(NUMBER_OF_CLUSTERS))
    available_clusters.remove(wrong_clusters[i])
    wrong_clusters[i] = random.choice(list(available_clusters))
wrong_cluster_model = ClusterEloModel(clusters=wrong_clusters)
wrong_cluster_simulator = Simulator(optimal_model, wrong_cluster_model, users, items, clusters, practice_length=PRACTICE_LENGTH)
wrong_cluster_simulator.simulate()
wrong_cluster_simulator.simulate_all()
```

```python
cluster_model = ClusterEloModel(clusters=clusters)
cluster_simulator = Simulator(optimal_model, cluster_model, users, items, clusters, practice_length=PRACTICE_LENGTH)
cluster_simulator.simulate()
cluster_simulator.simulate_all()
```

```python
naive_model = NaiveModel()
naive_simulator = Simulator(optimal_model, naive_model, users, items, clusters, practice_length=PRACTICE_LENGTH)
naive_simulator.simulate()
naive_simulator.simulate_all()
```

```python
const_model = ConstantModel(constant=TARGET_PROBABILITY)
const_simulator = Simulator(optimal_model, const_model, users, items, clusters, practice_length=PRACTICE_LENGTH)
const_simulator.simulate()
const_simulator.simulate_all()
```

```python
def intersections(plot, baseline, simulators):
    intersections_all = []
    subplot = plot.subplot(122)
    for simulator_name, simulator in simulators.iteritems():
        intersections = []
        for i in xrange(1, PRACTICE_LENGTH + 1):
            current_intersections = []
            for u in xrange(NUMBER_OF_USERS):
                current_intersections.append(len(
                    set(zip(*baseline._practice[u])[0][:i]) & set(zip(*simulator._practice[u])[0][:i])
                ))
            intersections.append(current_intersections)
        intersections_all.append(list(intersections))
        subplot.plot(range(PRACTICE_LENGTH), window_fun(intersections, numpy.mean, size=1), label=simulator_name)
    subplot.set_ylabel('Number of Items in Itersection of Practice')
    subplot.set_xlabel('Number of Attempts')  
    subplot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
    subplot = plot.subplot(121)    
    subplot.hist(map(lambda xs: xs[PRACTICE_LENGTH-1], intersections_all), bins=10)
    subplot.set_xlabel('Number of Items in Itersection of Practice')
    subplot.set_ylabel('Number of Users')
```

```python
figsize(17, 5)
intersections(plt, optimal_simulator, {
    'Elo': elo_simulator,
    'Elo, Clusters': cluster_simulator,
    'Elo, Clusters (wrong)': wrong_cluster_simulator,
    'Naive': naive_simulator,
    'Constant': const_simulator
})
```

```python
def precisions(plot, simulators):
    subplot = plot.subplot(121)
    for simulator_name, simulator in simulators.iteritems():
        simulator_predicted = []
        simulator_actual = []
        for u in xrange(NUMBER_OF_USERS):
            predictions = zip(*simulator._practice[u])[1]
            actual = zip(*simulator._practice[u])[2]
            simulator_predicted.extend(predictions)
            simulator_actual.extend(actual)
        subplot.set_xlabel('Number of Attempts')
        subplot.set_ylabel('RMSE')
            
    for simulator_name, simulator in simulators.iteritems():
        rmse_simulator = []
        for i in xrange(0, PRACTICE_LENGTH):
            predicted = []
            actual = []
            for u in xrange(NUMBER_OF_USERS):
                event = simulator._practice[u][i]
                predicted.append(event[1])
                actual.append(event[2])
            rmse_simulator.append(mean_squared_error(predicted, actual))
        means = numpy.array(running_fun(rmse_simulator, lambda mses: math.sqrt(numpy.mean(mses))))
        subplot.plot(range(PRACTICE_LENGTH), means, label='%s, %.4f' % (simulator_name, simulator.rmse(PRACTICE_LENGTH)))
    subplot = plot.subplot(122)
    for simulator_name, simulator in simulators.iteritems():
        rmse_simulator = []
        for u in xrange(NUMBER_OF_USERS):
            rmse_simulator.append(mean_squared_error(zip(*simulator._practice[u])[1], zip(*simulator._practice[u])[2]))
        means = numpy.array(running_fun(rmse_simulator, lambda mses: math.sqrt(numpy.mean(mses))))
        subplot.plot(range(NUMBER_OF_USERS), means, label='%s, %.4f' % (simulator_name, simulator.rmse(PRACTICE_LENGTH)))
        subplot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        subplot.set_xlabel('Number of Users')
        subplot.set_ylabel('RMSE')
```

```python
figsize(17, 5)
precisions(plt, {
    'Elo': elo_simulator,
    'Elo, Clusters': cluster_simulator,
    'Elo, Clusters (wrong)': wrong_cluster_simulator,
    'Naive': naive_simulator,
    'Constant': const_simulator,
    'Optimal': optimal_simulator
})
```

```python
def rmse_diff(plot, simulators):
    rmses = []
    rmses_all = []
    names = []
    for simulator_name, simulator in simulators.iteritems():
        names.append(simulator_name)
        rmses.append(simulator.rmse(PRACTICE_LENGTH))
        rmses_all.append(simulator.rmse_all(NUMBER_OF_ITEMS))
    index = numpy.arange(len(simulators))
    bar_width = 0.35
    plot.xticks(index + bar_width, names)
    plot.bar(index, rmses, bar_width, label='only a limited recommended practice', color=COLORS[0])
    plot.bar(index + bar_width, rmses_all, bar_width, label='all', color=COLORS[1])
    plot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plot.xlabel('Model')
    plot.ylabel('RMSE')
    plot.ylim(0.45, 0.57)
```

```python
rmse_diff(plt, {
    'Elo': elo_simulator,
    'Elo, Clusters': cluster_simulator,
    'Elo, Clusters (wrong)': wrong_cluster_simulator,
    'Naive': naive_simulator,
    'Constant': const_simulator,
    'Optimal': optimal_simulator
})
```
